```python
#Dataframe random
data = {'limit_balance': [20000.0,20000.0,240000.0,350000.0,10000.0],
       'age': [33.0,24.0,35.0,39.0,36.0],
       'pay_3': [0.0,2.0,0.0,0.0,-1.0],
       'pay_4': [0.0,2.0,0.0,0.0,-1.0],
       'bill_amt_3': [10510.0,322.0,181379.0,171127.0,3629.0],
       'bill_amt_4': [10472.0,322.0,170318.0,173283.0,6260.0],
       'pay_amt_3': [501.0,0.0,8000.0,6669.0,260.0],
       'pay_amt_4': [356.0,0.0,2500.0,5596.0,120.0],
       'material_status_new': ['married','single','single','married','single'],
       'sex_new': ['male','female','male','male','male'],
       'education_level_new': ['bachelors','bachelors','bachelors','bachelors','high school']}
```

```python
inf = pd.DataFrame(data,columns=data.keys())
inf
```

```python
# Menentukan persentase minimum pembayaran
minimum_payment_percent = 0.03

# Melakukan looping pada setiap baris data
for index, row in inf.iterrows():
    total_balance = row['limit_balance']
    minimum_payment = total_balance * minimum_payment_percent
    print(f"Minimum payment untuk {row['limit_balance']}: {minimum_payment}")

    pd.set_option('display.max_columns',None)
```

```python
full_pipeline = model_full_pipeline.transform(inf[['limit_balance',
 'age',
 'bill_amt_3',
 'bill_amt_4',
 'pay_amt_3',
 'pay_amt_4','material_status_new',
 'education_level_new',
 'pay_3',
 'pay_4',
 'sex_new']])
full_pipeline
```

```python
inf_final = full_pipeline.copy()
```
